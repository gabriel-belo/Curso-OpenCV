# Etapas para Criar Panoramas

1. Encontre os pontos-chave em todas as imagens
2. Encontre as correspondências entre pares
3. Estime as homografias entre pares
4. Refine as homografias
5. Una as imagens com mesclagem


**As imagens usadas para criar as panoâmicas precisam ser tiradas do mesmo ponto**

# Passos para criar panoramas usando OpenCV

1. Use a classe **sticher**

```python
import cv2
import glob
import matplotlib.pyplot as plt
import math
```

```python
# Read Images

imagefiles = glob.glob("boat/*")
imagefiles.sort()

images = []
for filename in imagefiles:
  img = cv2.imread(filename)
  img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
  images.append(img)

num_images = len(images)
```

```python
# Display Images
plt.figure(figsize=[30,10]) 
num_cols = 3
num_rows = math.ceil(num_images / num_cols)
for i in range(0, num_images):
  plt.subplot(num_rows, num_cols, i+1) 
  plt.axis('off')
  plt.imshow(images[i])
```

```python
# Stitch Images
stitcher = cv2.Stitcher_create()
status, result = stitcher.stitch(images)
if status == 0:
  plt.figure(figsize=[30,10]) 
  plt.imshow(result)
```
