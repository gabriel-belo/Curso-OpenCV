# <font style="color:black">Anotando Imagens</font>

Neste notebook, abordaremos como anotar imagens usando OpenCV. Aprenderemos a realizar as seguintes anotações em imagens:

* Desenhar linhas
* Desenhar círculos
* Desenhar retângulos
* Adicionar texto

Essas anotações são úteis quando você deseja anotar seus resultados para apresentações ou mostrar uma demonstração do seu aplicativo. As anotações também podem ser úteis durante o desenvolvimento e a depuração.

````python
# Import libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image 
%matplotlib inline
import matplotlib
matplotlib.rcParams['figure.figsize'] = (9.0, 9.0)
from IPython.display import Image
````

````python
# Read in an image
image = cv2.imread("Apollo_11_Launch.jpg", cv2.IMREAD_COLOR)

# Display the original image
plt.imshow(image[:,:,::-1])
````

## <font style="color:black">Desenhando uma Linha</font>

Vamos começar desenhando uma linha em uma imagem. Usaremos a função cv2.line para isso.

### <font style="color:rgb(8,133,37)">Sintaxe da Função</font>
``` python

img = cv2.line(img, pt1, pt2, color[, thickness[, lineType[, shift]]])
```

`img`: A imagem de saída que foi anotada.

A função possui **4 argumentos obrigatórios**:

1. `img`: Imagem na qual desenharemos uma linha
2. `pt1`: Primeiro ponto (coordenadas x, y) do segmento de linha
3. `pt2`: Segundo ponto do segmento de linha
4. `color`: Cor da linha que será desenhada

Outros argumentos opcionais importantes incluem:

1. `thickness`: Número inteiro que especifica a espessura da linha. O valor padrão é 1.
2. `lineType`: Tipo da linha. O valor padrão é 8, que representa uma linha de 8 segmentos. Normalmente, utiliza-se `cv2.LINE_AA` (linha suavizada ou com antisserrilhamento) para o tipo de linha.

### <font style="color:rgb(8,133,37)">Documentação do OpenCV</font>

**`line:`**https://docs.opencv.org/4.5.1/d6/d6e/group__imgproc__draw.html#ga7078a9fae8c7e7d13d24dac2520ae4a2

Vejamos um exemplo disso.

````python
imageLine = image.copy()

# The line starts from (200,100) and ends at (400,100)
# The color of the line is YELLOW (Recall that OpenCV uses BGR format)
# Thickness of line is 5px
# Linetype is cv2.LINE_AA

cv2.line(imageLine, (200, 100), (400, 100), (0, 255, 255), thickness=5, lineType=cv2.LINE_AA);

# Display the image
plt.imshow(imageLine[:,:,::-1])
````

## <font style="color:black">Drawing a Circle</font>


Let's start off by drawing a circle on an image. We will use cv2.circle function for this.

### <font style="color:rgb(8,133,37)">Functional syntx</font>
``` python
img = cv2.circle(img, center, radius, color[, thickness[, lineType[, shift]]])
```

`img`: The output image that has been annotated.

The function has **4 required arguments**:

1. `img`:    Image on which we will draw a line
2. `center`: Center of the circle
3. `radius`: Radius of the circle
4. `color`:  Color of the circle which will be drawn
    
Next, let's check out the (optional) arguments which we are going to use quite extensively.

1. `thickness`: Thickness of the circle outline (if positive). 
If a negative value is supplied for this argument, it will result in a filled circle.
2. `lineType`:  Type of the circle boundary. This is exact same as lineType argument in **cv2.line**
    
### <font style="color:rgb(8,133,37)">OpenCV Documentation</font>
    
**`circle:`** https://docs.opencv.org/4.5.1/d6/d6e/group__imgproc__draw.html#gaf10604b069374903dbd0f0488cb43670

Let's see an example of this.

````python
# Draw a circle
imageCircle = image.copy()

cv2.circle(imageCircle, (900,500), 100, (0, 0, 255), thickness=5, lineType=cv2.LINE_AA);

# Display the image
plt.imshow(imageCircle[:,:,::-1])
````

## <font style="color:black">Desenhando um Retângulo</font>

Usaremos a função **cv2.rectangle** para desenhar um retângulo em uma imagem. A sintaxe da função é a seguinte:

### <font style="color:rgb(8,133,37)">Sintaxe da função</font>

img = cv2.rectangle(img, pt1, pt2, color[, thickness[, lineType[, shift]]])

`img`: A imagem de saída que foi anotada.

A função possui **4 argumentos obrigatórios**:

1. `img`: Imagem na qual o retângulo será desenhado.

2. `pt1`: Vértice do retângulo. Normalmente, usamos o **vértice superior esquerdo** aqui.

3. `pt2`: Vértice do retângulo oposto a pt1. Normalmente, usamos o vértice **inferior direito** aqui.

4. `color`: Cor do retângulo

A seguir, vamos examinar os argumentos (opcionais) que usaremos com bastante frequência.

1. `thickness`: Espessura do contorno do círculo (se positivo).

Se um valor negativo for fornecido para este argumento, o resultado será um retângulo preenchido.

2. `lineType`: Tipo do contorno do círculo. Este argumento é exatamente igual ao argumento `lineType` em **cv2.line**

### <font style="color:rgb(8,133,37)">Links da Documentação do OpenCV</font>

**`rectangle:`**https://docs.opencv.org/4.5.1/d6/d6e/group__imgproc__draw.html#ga07d2f74cadcf8e305e810ce8eed13bc9

Vejamos um exemplo.

````python
# Draw a rectangle (thickness is a positive integer)
imageRectangle = image.copy()

cv2.rectangle(imageRectangle, (500, 100), (700,600), (255, 0, 255), thickness=5, lineType=cv2.LINE_8);

# Display the image
plt.imshow(imageRectangle[:,:,::-1])
````

## Adicionando Texto

Finalmente, vamos ver como podemos escrever um texto em uma imagem usando a função **cv2.putText**.

### Sintaxe funcional

img = cv2.putText(img, text, org, fontFace, fontScale, color[, thickness[, lineType[, bottomLeftOrigin]]])

`img`: A imagem de saída que foi anotada.

A função possui **6 argumentos obrigatórios**:

1. `img`: Imagem na qual o texto deve ser escrito.

2. `text`: Texto a ser escrito.

3. `org`: Canto inferior esquerdo do texto na imagem.

4. `fontFace`: Tipo de fonte
5. `fontScale`: Fator de escala da fonte, multiplicado pelo tamanho base específico da fonte.
6. `color`: Cor da fonte

Outros argumentos opcionais importantes incluem:

1. `thickness`: Número inteiro que especifica a espessura da linha do texto. O valor padrão é 1.
2. `lineType`: Tipo de linha. O valor padrão é 8, que representa uma linha de 8 colunas. Normalmente, usa-se `cv2.LINE_AA` (linha suavizada ou com antisserrilhamento) para o tipo de linha.

### <font style="color:rgb(8,133,37)">Documentação do OpenCV</font>

**`putText:`**https://docs.opencv.org/4.5.1/d6/d6e/group__imgproc__draw.html#ga5126f47f883d730f633d74f07456c576

Vejamos um exemplo disso.

````python
imageText = image.copy()
text = "Apollo 11 Saturn V Launch, July 16, 1969"
fontScale = 2.3
fontFace = cv2.FONT_HERSHEY_PLAIN
fontColor = (0, 255, 0)
fontThickness = 2

cv2.putText(imageText, text, (200, 700), fontFace, fontScale, fontColor, fontThickness, cv2.LINE_AA);

# Display the image
plt.imshow(imageText[:,:,::-1])
````
